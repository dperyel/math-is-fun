```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
dino = cv2.imread('./assets/dino.png')
dino = cv2.cvtColor(dino, cv2.COLOR_BGR2GRAY)
dino.shape
```

```python
dino_s = cv2.resize(dino, (0,0), fx=0.3, fy=0.3)
dino_s.shape
```

```python
plt.figure(figsize=(10,10))

plt.imshow(dino_s, cmap='gray', interpolation='nearest')

plt.xticks(np.arange(0.5,30), None)
plt.yticks(np.arange(0.5,30), None)
plt.grid(color="#999999")
plt.show()
```

```python
plt.figure(figsize=(10,10))

rotation_matrix = cv2.getRotationMatrix2D((15, 15), 45, 1)
dino_rot = cv2.warpAffine(dino_s, rotation_matrix, (30, 30))
plt.imshow(dino_rot, cmap='gray', interpolation='nearest')

plt.xticks(np.arange(0.5,30), None)
plt.yticks(np.arange(0.5,30), None)
plt.grid(color="#999999")
plt.show()
```

```python
print(rotation_matrix)
```

```python
plt.figure(figsize=(10,10))

dino_gaus = cv2.GaussianBlur(dino_s, (3,3), cv2.BORDER_DEFAULT)

plt.imshow(dino_gaus, cmap='gray', interpolation='nearest')

plt.xticks(np.arange(0.5,30), None)
plt.yticks(np.arange(0.5,30), None)
plt.grid(color="#999999")
plt.show()
```

```python
plt.figure(figsize=(10,10))

dino_sobel = cv2.Sobel(dino_gaus, cv2.CV_8U, 1, 0, ksize=3)

plt.imshow(dino_sobel, cmap='gray', interpolation='nearest')

plt.xticks(np.arange(0.5,30), None)
plt.yticks(np.arange(0.5,30), None)
plt.grid(color="#999999")
plt.show()
```

```python
plt.figure(figsize=(10,10))

dino_canny = cv2.Canny(dino_gaus, threshold1=200, threshold2=255)

plt.imshow(dino_canny, cmap='gray', interpolation='nearest')

plt.xticks(np.arange(0.5,30), None)
plt.yticks(np.arange(0.5,30), None)
plt.grid(color="#999999")
plt.show()
```
